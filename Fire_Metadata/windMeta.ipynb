```python
for i, row in fireDF.iterrows():
    fire = row.fireID
    print(fire, row.fire_start)
    wspd_file = f'{root_dir}/{fire}/{fire}_windspeed.nc'
    wdir_file = f'{root_dir}/{fire}/{fire}_winddir.nc'
    with Dataset(wspd_file,'r') as nc_file:
        time_var = nc_file.variables['time']
        raw_times = time_var[:]
        wspd_dates = num2date(raw_times, units=time_var.units)
    with Dataset(wdir_file,'r') as nc_file:
        time_var = nc_file.variables['time']
        raw_times = time_var[:]
        wdir_dates = num2date(raw_times, units=time_var.units)
    for i in [0,23,24]:
        print(wspd_dates[i].strftime('%Y-%m-%d %H:%M:%S'), wdir_dates[i].strftime('%Y-%m-%d %H:%M:%S'),raw_times[i])
    print()
    
```

```python
fireDF['wind_start'] = ''
for i, row in fireDF.iterrows():
    fire = row.fireID
    wspd_file = f'{root_dir}/{fire}/{fire}_windspeed.nc'
    with Dataset(wspd_file,'r') as nc_file:
        time_var = nc_file.variables['time']
        raw_times = time_var[:]
        wspd_dates = num2date(raw_times, units=time_var.units)
        fireDF.loc[i,'wind_start'] = wspd_dates[i].strftime('%Y-%m-%d %H:%M:%S')
fireDF.wind_start = pd.to_datetime(fireDF.wind_start)
fireDF 
```
