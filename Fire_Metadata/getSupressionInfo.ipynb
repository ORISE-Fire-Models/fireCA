```python
import pandas as pd
import numpy as np
import os
from datetime import datetime

pd.set_option('display.max_columns', None)
```

```python
wdir = "data/ics209plus-wildfire/"
#vdir = "/mnt/DataDrive1/data/zholden/VIIRS/"
```

```python
# Load files
flist = sorted([os.path.join(wdir, f) for f in os.listdir(wdir) if f.endswith(".csv")])
flist
```

```python
all_dfs = []
for file_path in flist:
    try:
        # Read each CSV file into a DataFrame
        df = pd.read_csv(file_path)
        # Add a column for the original file name (optional but often useful)
        df['source_file'] = os.path.basename(file_path)
        all_dfs.append(df)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
```

```python
df = pd.concat(all_dfs, ignore_index=True)
df
```

```python
df = df.sort_index(axis=1)
df
```

```python
with open('data/fire_list.txt', "r") as file:
    fireIDs = file.readlines()
for i in range(len(fireIDs)):
    fireIDs[i] = fireIDs[i].strip()
sorted(fireIDs)
```

```python
limitRows = []

for fire in fireIDs:
    fireDF = df.loc[df.MTBS_FIRE_LIST.str.contains(fire, case=False, na=False),['MTBS_FIRE_LIST','WF_PEAK_AERIAL_DATE','WF_PEAK_PERSONNEL_DATE']]
    fireDF['WF_PEAK_AERIAL_DATE'] = pd.to_datetime(fireDF['WF_PEAK_AERIAL_DATE'])
    fireDF['WF_PEAK_PERSONNEL_DATE'] = pd.to_datetime(fireDF['WF_PEAK_PERSONNEL_DATE'])
    mtbsList = fireDF.MTBS_FIRE_LIST.iloc[0].translate({ord(i): None for i in "[]''"}).split(',')
    for s in fireDF.MTBS_FIRE_LIST.iloc[0].translate({ord(i): None for i in "[]''"}).split(','):
        if fire in s:
            name = s.split('-')[-1].strip().replace(' ','_')
    indices = str(list(fireDF.index))
    limitRows.append({'fireID':fire, 
                      'fire_name':name, 
                      'peak_aerial': fireDF['WF_PEAK_AERIAL_DATE'].min(),
                      'peak_personnel': fireDF['WF_PEAK_PERSONNEL_DATE'].min(),
                      'fs209_indices': indices
                     })
    #print(fire, limitRow)
limitDF =  pd.DataFrame(limitRows)
limitDF['first_peak'] = np.minimum(limitDF.peak_aerial,limitDF.peak_personnel)
```

```python
fireInfo = pd.read_csv('data/fire_info.csv')
fireInfo = fireInfo[['fireID','fire_name','fire_start','hours_of_data','days_of_data',
                    'peak_aerial', 'peak_personnel','first_supression','days_til_sup','fs209_indices']]
fireInfo = fireInfo.rename(columns={'days_til_sup':'days_til_first', 'first_supression':'first_suppression'})
for col in ['fire_start','peak_aerial','peak_personnel','first_suppression']:
    fireInfo[col] = pd.to_datetime(fireInfo[col])
fireInfo
```

```python
fireInfo['hours_til_first'] = ((fireInfo.first_suppression - fireInfo.fire_start).dt.total_seconds()/60/60).round().astype('int')
fireInfo['hours_til_aerial'] = ((fireInfo.peak_aerial - fireInfo.fire_start).dt.total_seconds()/60/60).round().astype('int')
fireInfo['hours_til_personnel'] = ((fireInfo.peak_personnel - fireInfo.fire_start).dt.total_seconds()/60/60).round().astype('int')
fireInfo['days_til_first'] = ((fireInfo.hours_til_first)/24).round().astype('int')
fireInfo['days_til_aerial'] = ((fireInfo.hours_til_aerial)/24).round().astype('int')
fireInfo['days_til_personnel'] = ((fireInfo.hours_til_personnel)/24).round().astype('int')
fireInfo
```
